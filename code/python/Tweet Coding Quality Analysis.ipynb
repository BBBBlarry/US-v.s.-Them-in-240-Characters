```python
import json
import numpy as np
import pandas as pd
from sklearn.metrics import precision_score, recall_score, f1_score
```

## Generate

```python
tweet_json_file = "../../data/tweets/all_tweets_with_coding.json"
with open(tweet_json_file, 'r') as f:
    tweets = json.load(f)
```

```python
num_samples = 300

np.random.seed(2)

random_sample = np.random.choice(tweets, num_samples)
```

```python
for sample in random_sample:
    print(sample["text"])
    print("is_attack", sample["is_attack"])
    print("is_policy", sample["is_policy"])
    print("is_contrast", sample["is_contrast"])
    print()
```

```python
validation_file = "../../data/validation/sample_validation_init.csv"
with open(validation_file, 'w') as f:
    f.write("{}, {}, {}, {}, {}, {}, {}\n".format("text", "is_attack", "is_policy", "is_contrast", "is_attack_correct", "is_policy_correct", "is_contrast_correct"))
    for sample in list(random_sample):
        text = sample["text"]
        text = text.replace(',', ';')
        text = text.replace('\n', ' ')
        is_attack = sample["is_attack"]
        is_policy = sample["is_policy"]
        is_contrast = sample["is_contrast"]
        f.write("\"{}\", {}, {}, {}, 1, 1, 1\n".format(text, is_attack, is_policy, is_contrast))
```

## Validate

```python
validated_file = "../../data/validation/sample_validation.csv"
validated_df = pd.read_csv(validated_file)
validated_df.is_attack = validated_df.is_attack == " True"
validated_df.is_policy = validated_df.is_policy == " True"
validated_df.is_contrast = validated_df.is_contrast == " True"

validated_df.is_attack = validated_df.is_attack.astype(int)
validated_df.is_policy = validated_df.is_policy.astype(int)
validated_df.is_contrast = validated_df.is_contrast.astype(int)

# construct gold labels
validated_df["is_attack_gold"] = validated_df.is_attack == validated_df.is_attack_correct
validated_df["is_policy_gold"] = validated_df.is_policy == validated_df.is_policy_correct
validated_df["is_contrast_gold"] = validated_df.is_contrast == validated_df.is_contrast_correct

validated_df.is_attack_gold = validated_df.is_attack_gold.astype(int)
validated_df.is_policy_gold = validated_df.is_policy_gold.astype(int)
validated_df.is_contrast_gold = validated_df.is_contrast_gold.astype(int)
```

### Validation Stats

```python
attack_p = precision_score(validated_df.is_attack_gold, validated_df.is_attack)
attack_r = recall_score(validated_df.is_attack_gold, validated_df.is_attack)
attack_f1 = f1_score(validated_df.is_attack_gold, validated_df.is_attack)
print("Attacks: Precision {}, Recall {}, F1 {}".format(attack_p, attack_r, attack_f1))
```

```python
policy_p = precision_score(validated_df.is_policy_gold, validated_df.is_policy)
policy_r = recall_score(validated_df.is_policy_gold, validated_df.is_policy)
policy_f1 = f1_score(validated_df.is_policy_gold, validated_df.is_policy)
print("Policies: Precision {}, Recall {}, F1 {}".format(policy_p, policy_r, policy_f1))
```

```python
contrast_p = precision_score(validated_df.is_contrast_gold, validated_df.is_contrast)
contrast_r = recall_score(validated_df.is_contrast_gold, validated_df.is_contrast)
contrast_f1 = f1_score(validated_df.is_contrast_gold, validated_df.is_contrast)
print("Contrasts: Precision {}, Recall {}, F1 {}".format(contrast_p, contrast_r, contrast_f1))
```
