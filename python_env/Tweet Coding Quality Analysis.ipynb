```python
import json
import numpy as np
```

```python
tweet_json_file = "./data/all_tweets_with_coding.json"
with open(tweet_json_file, 'r') as f:
    tweets = json.load(f)
```

```python
num_samples = 150
attack_samples = 150

attack_tweets = [tweet for tweet in tweets if tweet["is_attack"]]

np.random.seed(42)

random_total_sample = np.random.choice(tweets, num_samples)
random_attack_sample = np.random.choice(attack_tweets, num_samples)

random_sample = list(random_total_sample) + list(random_attack_sample)
```

```python
for sample in random_attack_sample:
    print(sample["text"])
    print("is_attack", sample["is_attack"])
    print("is_policy", sample["is_policy"])
    print("is_contrast", sample["is_contrast"])
    print()
```

```python
validation_file = "sample_validation_init.csv"
with open(validation_file, 'w') as f:
    f.write("{}, {}, {}, {}\n".format("text", "is_attack", "is_policy", "is_contrast"))
    for sample in list(random_total_sample) + list(random_attack_sample):
        text = sample["text"]
        text = text.replace(',', ';')
        text = text.replace('\n', ' ')
        is_attack = sample["is_attack"]
        is_policy = sample["is_policy"]
        is_contrast = sample["is_contrast"]
        f.write("\"{}\", {}, {}, {}\n".format(text, is_attack, is_policy, is_contrast))
```
